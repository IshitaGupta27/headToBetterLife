#  Data Scrapping -- Books

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
Book_Info=[]
Url_Info=[]
Abstract_Info=[]
output_rows =pd.DataFrame(columns = ['BOOK_NAME', 'ABSTRACT','URL'])
for i in range(0,140,20):
    url="https://www.smashwords.com/books/category/511/newest/0/free/any/"+str(i)
    page=requests.get(url)
    soup=BeautifulSoup(page.content,"html.parser")
    for link in soup.find_all(class_='library-title'):
        Book_Info.append(link.get_text())
        Url_Info.append(link.get('href'))
    for linkp in soup.find_all(class_='card bg-light border-light p-2'):
        Abstract_Info.append(linkp.get_text())
    
```

```python
ABSTRACT = [item.strip() for item in Abstract_Info if str(item)]
output_rows['BOOK_NAME'] = Book_Info
output_rows['ABSTRACT']=ABSTRACT
output_rows['URL']=Url_Info
```

```python
output_rows.to_csv('BOOKS.csv')
```

```python
!pip install selenium
import csv
#Importing packages
from selenium import webdriver
import pandas as pd
driver = webdriver.Chrome('C:\Program Files\chromedriver_win32\chromedriver.exe')
driver.get('https://adaa.org/helpful-inspiring-songs')

first_part="//*[@id='block-bootstrap-sass-content']/div/div/div/ul[1]/li["
second_part="]/a"

file=open("product.csv","wt",newline='')
writer=csv.writer(file)

song_name=[]
Url=[]
output =pd.DataFrame(columns = ['SONG_NAME', 'URL'])
for n in range(1,28):
        final_path=first_part+str(n)+second_part
        table_data= driver.find_element_by_xpath(final_path)
        song_name.append(table_data.text)
        Url.append(table_data.get_attribute('href'))
        
        
output['SONG_NAME'] = song_name
output['URL']=Url
output.to_csv('SONGS.csv')
```
